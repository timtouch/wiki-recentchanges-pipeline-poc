```sql
-- Record the current Bronze row count
SELECT COUNT(*) AS bronze_before FROM wiki_poc.poc.bronze_recentchange_raw;
```

```sql
-- Should show continuous rows with no multi-minute gap
SELECT
  from_unixtime(FLOOR(unix_timestamp(ingest_timestamp) / 60) * 60) AS minute,
  COUNT(*) AS rows_landed
FROM wiki_poc.poc.bronze_recentchange_raw
WHERE ingest_timestamp >= NOW() - INTERVAL 30 MINUTES
GROUP BY 1
ORDER BY 1;
```
